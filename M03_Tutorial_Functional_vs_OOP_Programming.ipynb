```python
#User function template for Python

class Solution:    
    def binarysearch(self, arr, n, k):
        # code here
        
        """
            This binary search function was adapted for use in Python 3 from this pseudocode from 
            https://en.wikipedia.org/wiki/Binary_search_algorithm.
            
            function binary_search(A, n, T) is
                L := 0
                R := n − 1
                while L ≤ R do
                    m := floor((L + R) / 2)
                    if A[m] < T then
                        L := m + 1
                    else if A[m] > T then
                        R := m − 1
                    else:
                        return m
                return unsuccessful
            
            Basically, the middle value is low_value plus high_value floor divided by 2. The binary search operates by 
            cutting the search space of the sorted array in half for each iteration of the while loop until it finds the 
            correct answer (i.e. return middle_value) or ends in failure to find the correct value (i.e. returns -1). 
        """
        low_value = 0
        high_value = n - 1
        
        while low_value <= high_value:
            middle_value = (low_value + high_value) // 2
            if arr[middle_value] < k:
                low_value = middle_value + 1
            elif arr[middle_value] > k:
                high_value = middle_value - 1
            else:
                return middle_value
        return -1 
```
